```python
# Getting a list of countries where English is an official language.

#Loading requests module and BeautifulSoup library
import requests, bs4

#Using requests to connect to a URL and get data from it 
url = "https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language"

#Creating a BeautifulSoup object 
html_data = requests.get(url)

#Getting attributes of BeautifulSoup object(i.e. HTML elements)
bs4_object = bs4.BeautifulSoup(html_data.text, "html5lib")

#Getting data without HTML.
table_data1 = bs4_object.select(".mw-parser-output table:nth-of-type(1) > tbody > tr > td:nth-child(2) > a")
languages1 = [str(str(x).split('">')[1]).split('</a>')[0] for x in table_data1]
table_data2 = bs4_object.select(".mw-parser-output table:nth-of-type(2) > tbody > tr > td:nth-child(2) > a")
languages2 = [str(str(x).split('">')[1]).split('</a>')[0] for x in table_data2]
table_data3 = bs4_object.select(".mw-parser-output table:nth-of-type(3) > tbody > tr > td:nth-child(2) > a")
languages3 = [str(str(x).split('">')[1]).split('</a>')[0] for x in table_data3]
languages = languages1 + languages2 + languages3
print(languages)
```

```python
# Getting a list of countries with country codes

#Using requests to connect to a URL and get data from it 
url_code = "https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3"

#Creating a BeautifulSoup object 
html_data_code = requests.get(url_code)

#Getting attributes of BeautifulSoup object(i.e. HTML elements)
bs4_object_code = bs4.BeautifulSoup(html_data_code.text, "html5lib")

#Getting data without HTML.
table_data1_code = bs4_object_code.select(".plainlist > ul > li > span")
country_codes = [str(str(x).split('">')[1]).split('</span>')[0] for x in table_data1_code]

#Using requests to connect to a URL and get data from it 
url_full = "https://en.wikipedia.org/wiki/ISO_3166-1_alpha-3"

#Creating a BeautifulSoup object 
html_data_full = requests.get(url_full)

#Getting attributes of BeautifulSoup object(i.e. HTML elements)
bs4_object_full = bs4.BeautifulSoup(html_data_full.text, "html5lib")

#Getting data without HTML.
table_data1_full = bs4_object_full.select(".plainlist > ul > li > a:nth-of-type(1) ")
country_full = [str(str(x).split('">')[1]).split('</a>')[0] for x in table_data1_full]

our_keys = ["country_code", "country_name"]
our_values = list((country_codes, country_full))
#res = {our_keys[i]: our_values[i] for i in range(len(our_keys))} 

country_list_code = dict(zip(our_keys, our_values))

country = pd.DataFrame(country_list_code)
country = country.replace(to_replace ="United States of America", value ="United States")
country = country.replace(to_replace ="United Kingdom of Great Britain and Northern Ireland", value ="United Kingdom")
```

```python
# Loading mapping data into dataframe mapping
mapping = pd.read_csv("mapping.csv", encoding = "ISO-8859-1")

mapping.head()
```

```python
# details of dataframe mapping
mapping.info()
```

```python
#handling missing data
mapping = mapping.drop("Blanks", axis = 1) #removing column without category list
mapping = mapping[~mapping.isnull().any(axis=1)] #removing row with null value
mapping.info()
```

```python
# converting common column to lowercase
mapping['category_list'] = mapping['category_list'].str.lower()

# replacing '0' with 'na'
mapping['category_list'] = mapping['category_list'].apply(lambda x: x.replace('0', 'na'))
print(mapping['category_list'])
```

```python
# checking if there is any duplicate value in category_list
mapping.category_list.duplicated().any()
```

```python
# unpivoting data of mapping
map1 = mapping.melt(id_vars = ["category_list"], value_vars = ["Automotive & Sports", "Cleantech / Semiconductors", "Entertainment", "Health", "Manufacturing", "News, Search and Messaging", "Others", "Social, Finance, Analytics, Advertising"], var_name="main_sector")
map1 = map1[map1["value"] == 1]

map1 = map1.drop("value", axis = 1) #dropping column value
map1 = map1.rename(columns = {"category_list" : "map_category_list"}) #renaming column category_list
map1["main_sector"] = map1["main_sector"].str.lower() #converting value of main_sector to lowercase
map1["map_category_list"] = map1["map_category_list"].str.lower() #converting value of map_category_list to lowercase
map1.head()
```

```python
# merging dataframe map1 to master dataframe master_frame
master_frame = pd.merge(master_frame, map1, how = "inner", left_on = "primary_sector", right_on = "map_category_list")
master_frame = master_frame.drop("map_category_list", axis = 1) #dropping column map_category_list
master_frame = master_frame[master_frame["main_sector"].isin(map1["main_sector"].tolist())] # taking 8 main sectors

master_frame.head()
```

```python
# groupby for column main_sector of dataframe d2
grp_inv2 = d2.groupby("main_sector")

#total number of investments for each main sector in a column number_of_investments
count_investment2  = grp_inv2["raised_amount_usd"].count()
main_inv_count2 = pd.DataFrame(count_investment2)
main_inv_count2 = main_inv_count2.rename(columns = {"raised_amount_usd" : "number_of_investments"})
d2 = pd.merge(d2, main_inv_count2, how = "inner", on = "main_sector")

#total amount invested in each main sector in a column total_investments
sum_investment2  = grp_inv2["raised_amount_usd"].sum()
main_inv_sum2 = pd.DataFrame(sum_investment2)
main_inv_sum2 = main_inv_sum2.rename(columns = {"raised_amount_usd" : "total_investments"})
d2 = pd.merge(d2, main_inv_sum2, how = "inner", on = "main_sector")
d2
```
